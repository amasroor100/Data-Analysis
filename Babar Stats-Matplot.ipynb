```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
babar=pd.read_csv("babar.csv")
```

```python
babar
```

```python
runs = babar["Runs"]
```

```python
runs
```

```python
y=runs.sum()
```

```python
babar["Runs %"]=(runs/y)*100
```

```python
babar
```

```python
babar
```

```python
plt.bar(babar["Opposition"], babar["Runs %"])

plt.xticks(rotation=45)

plt.show
```

```python
x= babar["Opposition"]
y=babar["Runs"]
fig , ax = plt.subplots(figsize = (10,5))
plt.barh(x,y)
pos = ["right" , 'left' , 'top' , 'bottom']
for i in pos:
    ax.spines[i].set_visible(False)
    
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')

ax.grid(visible = True , color = 'grey' , alpha = 0.2)
ax.invert_yaxis()
ax.set_title("Babar Azam Total Runs" , loc ='center' , fontdict = {'fontsize' : 20})

plt.show
```
